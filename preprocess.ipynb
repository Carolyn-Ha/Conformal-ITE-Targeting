### 연령
- 만 20-34세로 filtering
- 더미 변수로: 20-24세 / 25-29세 / 30-34세

```python
df_selected = df_selected[(df_selected["X1 - 연령"] >= 20) & (df_selected["X1 - 연령"] <= 34)]
df_selected["X1 - 연령"].value_counts()
```

```python
# 1단계: 구간화 함수 정의
def categorize_age(age):
    if 20 <= age <= 24:
        return "20_24"
    elif 25 <= age <= 29:
        return "25_29"
    elif 30 <= age <= 34:
        return "30_34"
    else:
        return np.nan

# 2단계: 연령대 범주 변수 생성
df_selected["X1 - 연령대"] = df_selected["X1 - 연령"].apply(categorize_age)

# 3단계: 더미 변수 생성 (원하는 경우 drop_first=True 가능)
# df_selected = pd.get_dummies(df_selected, columns=["연령대"], prefix="X1 - 연령", dtype=int)

# 연령대를 범주형(factor) 변수로 지정 - ordinal
df_selected["X1 - 연령대"] = pd.Categorical(
    df_selected["X1 - 연령대"],
    categories=["20_24", "25_29", "30_34"],
    ordered=True
)
```

### 부모님의 최종 학력

- -1(모름/무응답), 1(무학), 2(초등학교), 3(중학교), 4(고등학교), 5(2-3년제 대학(초급대)), 6(4년제 대학), 7(대학원) 

```python
df_selected["X1 - 아버님의 최종 학력"].value_counts()
df_selected["X1 - 어머님의 최종 학력"].value_counts()
```

```python
# 학력 코드 → 연한 매핑
edu_year_map = {
    -1: np.nan,  # 모름/무응답
    1: 0,
    2: 6,
    3: 9,
    4: 12,
    5: 14,
    6: 16,
    7: 18
}

# 각 부모의 학력을 연한으로 매핑
father_years = df_selected["X1 - 아버님의 최종 학력"].map(edu_year_map)
mother_years = df_selected["X1 - 어머님의 최종 학력"].map(edu_year_map)

# 평균 학력 연한 계산 (NaN 제외한 평균)
df_selected["X1 - 부모학력"] = pd.concat([father_years, mother_years], axis=1).mean(axis=1, skipna=True)
```

### 부모님 월평균 소득
- 1(100만원 미만)
- 2(100~200만원 미만)
- 3(200~300만원 미만)
- 4(300~400만원 미만)
- 5(400~500만원 미만)
- 6(500~700만원 미만)
- 7(700~1,000만원 미만)
- 8(1,000만원 이상)
- 9(소득없음)
- 10(안계심(사망))

- 더미변수로: 300만원 미만 / 300-500만원 / 500만원 이상

```python
# 단계 1: 구간 분류 함수 정의
def categorize_income(code):
    if code in [1, 2, 3]:
        return "저소득"
    elif code in [4, 5]:
        return "중소득"
    elif code in [6, 7, 8]:
        return "고소득"
    else:
        return np.nan  # 9 (소득없음), 10 (사망) → 결측 처리

# 단계 2: 새 변수 생성
df_selected["X1 - 부모소득구간"] = df_selected["X1 - 대학 입학(편입) 당시 부모님의 월평균 소득"].apply(categorize_income)

# 단계 3: 더미 변수로 변환 (원한다면 drop_first=True로 기준 더미 제거 가능)
# df_selected = pd.get_dummies(df_selected, columns=["부모소득구간"], prefix="X1 - 부모소득", dtype=int)

# Categorical(ordinal) 변수 생성
df_selected["X1 - 부모소득구간"] = pd.Categorical(
    df_selected["X1 - 부모소득구간"],
    categories=["저소득", "중소득", "고소득"],
    ordered=True  # 순서형으로 하고 싶지 않다면 False로
)
```
